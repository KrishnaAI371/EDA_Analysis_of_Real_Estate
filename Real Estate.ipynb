# Univariate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["House_Age"], bins=20, kde=True)
plt.title("Distribution of House Age")
plt.xlabel("House_Age (Years)")
plt.ylabel("Count")
plt.show()
```

## Price per Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Price_Per_Unit'],bins =20, kde=True)
plt.title('Distribution Of House Price per unit Area')
plt.xlabel('Price_Per_Unit')
plt.ylabel('Count')
plt.show()
```

# Bivariate Analysis

### Price vs House Age

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='House_Age',y ='Price_Per_Unit', data = df)
plt.title('Price vs House_Age')
plt.show()
```

## Price vs Distance to MRT

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Distance_Mrt", y="Price_Per_Unit", data=df)
plt.title("Price vs. Distance to MRT")
plt.show()
```

## Price vs Convenience stores

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Convenience_Stores', y ='Price_Per_Unit', data = df)
plt.title('Price vs Convenience_Stores')
plt.show()
```

## Correlation Heatmap

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap = 'coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# Key Insights

### 1) House price decreases slightly with age.
### 2) Properties closer to MRT stations tends to have higher prices.
### 3) Most Covenience Stores nearby generally increases property values
### 4) Longitude and Latitude also show realtion with house price that why location  matters.


# Conclusion

### Proximity to MRT stations, house age, and nearby amenities significantly influence real estate prices in Taiwan.
